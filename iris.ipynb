```python
# Dependencies and set up
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
# Load the iris dataset into an array
iris = load_iris()
```

```python
# Load the iris array into a dataframe
iris_data = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
# Create feature for the numerical value of class
iris_data['class'] = iris.target
```

```python
# Create feature for the descriptive value of class
iris_data['class_name'] = np.where(iris_data['class'] == 0, iris.target_names[0], np.where(iris_data['class'] == 1, iris.target_names[1], np.where(iris_data['class'] == 2, iris.target_names[2], 'error')))
```

```python
# Investigate if there are any N/A values
iris_data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']].isna().sum()
```

```python
# Look at summary statistics for the four numerical columns
iris_data.iloc[:,0:4].describe()
```

```python
# Create a pairplot to generally show the of the variables
sns.pairplot(iris_data, vars = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'], hue = 'class_name', palette = 'Set2')
```
